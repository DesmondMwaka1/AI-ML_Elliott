```python
#a Tensor is the fundamental data structure used to store and manipulate data
#it is a multi-dimensional array of numbers.
a=torch.tensor([2,5,1])
b=torch.tensor([6,9,3])

sum=a+b
mal=a*b
minus=b-a
```

```python
#for matrices
c=torch.tensor([[6,3,8],[2,7,1]])
d=torch.tensor([[0,1,5],[9,3,4],[2,6,1]])
mal1=torch.mm(c,d)
mal1
```

```python
#for vectors to get dot product
e=torch.tensor([3,1,6,7,3])
f=torch.tensor([7,3,8,1,0])
sum1=torch.dot(e,f)
sum1
```
