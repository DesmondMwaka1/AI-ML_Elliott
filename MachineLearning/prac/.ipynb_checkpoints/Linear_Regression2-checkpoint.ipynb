```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv(r"FuelConsumption.csv")
```

```python
df.head(5)
```

```python
mask=np.random.rand(len(df))<0.8
train=df[mask]
test=df[~mask]
```

```python
x1=train.FUELCONSUMPTION_CITY
y1=train.CO2EMISSIONS
```

```python
plt.scatter(x1,y1,color='purple')
plt.xlabel("FUELCONSUMPTION_CITY")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

```python
#train_model
from sklearn import linear_model
regres=linear_model.LinearRegression()
trainX=np.asanyarray(train[["FUELCONSUMPTION_CITY"]])
trainY=np.asanyarray(train[["CO2EMISSIONS"]])
regres.fit(trainX,trainY)
print("Coefficient:",regres.coef_)
print("X-Intercept:",regres.intercept_)
```

```python
#plot model
plt.scatter(train.FUELCONSUMPTION_CITY,train.CO2EMISSIONS,color="blue")
plt.plot(train.FUELCONSUMPTION_CITY,regres.coef_[0][0]*train.FUELCONSUMPTION_CITY+regres.intercept_,color="purple")
plt.xlabel("FUELCONSUMPTION_CITY")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

```python
#Error
from sklearn.metrics import r2_score
testX=np.asanyarray(test[["FUELCONSUMPTION_CITY"]])
testY=np.asanyarray(test[["CO2EMISSIONS"]])
predY=regres.predict(testY)
print(np.mean(np.absolute(testY-predY)))
print(r2_score(predY,testY))
```

```python
#plot all
plt.scatter(train.FUELCONSUMPTION_CITY,train.CO2EMISSIONS,color="blue")
plt.scatter(test.FUELCONSUMPTION_CITY,test.CO2EMISSIONS,color="red")
plt.plot(train.FUELCONSUMPTION_CITY,regres.coef_[0][0]*train.FUELCONSUMPTION_CITY+regres.intercept_,color="purple")
plt.xlabel("FUELCONSUMPTION_CITY")
plt.ylabel("CO2EMISSIONS")
plt.show()
```
