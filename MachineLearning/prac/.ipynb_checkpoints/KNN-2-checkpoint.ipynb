```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing

df=pd.read_csv(r"teleCust1000t.csv")
df.head(5)
```

```python
x=df[["region","tenure","age","marital","address","income","ed","employ","retire","gender","reside"]].values #2 dimensional
y=df["custcat"].values #1 dimensional

x=preprocessing.StandardScaler().fit(x).transform(x.astype(float))
```

```python
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2,random_state=4)
print ('Train set:', trainx.shape,  trainy.shape)
print ('Test set:', testx.shape,  testy.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=4
model1=KNeighborsClassifier(n_neighbors=k).fit(trainx,trainy)
model1
```

```python
predy=model1.predict(testx)
predy[1:25]
```

```python
from sklearn import metrics
print("Train set accuracy:",metrics.accuracy_score(trainy,model1.predict(trainx)))
print("Test set accuracy:",metrics.accuracy_score(testy,predy))
```
