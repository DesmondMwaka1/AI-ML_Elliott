```python
import sys
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree

df=pd.read_csv(r"drugs.csv")
df.head(5)
```

```python
x=df[["Age","Sex","BP","Cholesterol","Na_to_K"]].values
x[1]
```

```python
from sklearn import preprocessing
df_sex=preprocessing.LabelEncoder()
df_sex.fit(['F','M'])
x[:,1]=df_sex.transform(x[:,1])

df_BP=preprocessing.LabelEncoder()
df_BP.fit(['LOW','NORMAL','HIGH'])
x[:,2]=df_BP.transform(x[:,2])

df_Cholesterol=preprocessing.LabelEncoder()
df_Cholesterol.fit(['NORMAL','HIGH'])
x[:,3]=df_Cholesterol.transform(x[:,3])

```

```python
x[3:17]
```

```python
y = df["Drug"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree # it shows the default parameters
```

```python
drugTree.fit(trainx,trainy)
```

```python
predy=drugTree.predict(testx)
print("Predictions:",predy[4:8])
print("Actual:",testy[4:8])
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(testy,predy))
```
