```python

# Assuming df is already loaded and these variables are correctly defined
x = df.index
m = df["Engine size"]
n = df["Cylinders"]
a = df["Fuel consumption comb"]
b = df["CO@ emissions"]

# 1. Create a figure with a grid of 4 rows and 1 column
fig, axes = pt.subplots(nrows=4, ncols=1, figsize=(10, 12))

# 2. Plot each relationship on its own axis (ax[i])

# Plot 1: Engine Size
axes[0].scatter(x, m, s=10, c='blue')
axes[0].set_title('Engine Size vs. Index')
axes[0].set_ylabel('Engine Size')
axes[0].grid(axis='y', alpha=0.5)

# Plot 2: Cylinders
axes[1].scatter(x, n, s=10, c='red')
axes[1].set_title('Cylinders vs. Index')
axes[1].set_ylabel('Cylinders')
axes[1].grid(axis='y', alpha=0.5)

# Plot 3: Fuel Consumption
axes[2].scatter(x, a, s=10, c='green')
axes[2].set_title('Fuel Consumption vs. Index')
axes[2].set_ylabel('Fuel Consumption (L/100km)')
axes[2].grid(axis='y', alpha=0.5)

# Plot 4: CO2 Emissions
axes[3].scatter(x, b, s=10, c='purple')
axes[3].set_title('CO2 Emissions vs. Index')
axes[3].set_ylabel('CO2 Emissions (g/km)')
axes[3].set_xlabel('Record Index') # Only label the bottom x-axis
axes[3].grid(axis='y', alpha=0.5)

# 3. Automatically adjust subplot parameters for a tight layout
pt.tight_layout()
pt.show()
```
