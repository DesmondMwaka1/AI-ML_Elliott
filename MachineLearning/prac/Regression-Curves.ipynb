```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df=pd.read_csv(r"polynomial_curves_data.csv")
df[0:len(df)]
```

```python
plt.scatter(df["X"],df["Linear_Y"],color="blue")
plt.xlabel("X")
plt.ylabel("Linear_Y")
plt.show()
```

```python
#linear Y = mX+c
slope=0.3
intercept=5

x_vals_1=np.linspace(-5,5,500) #500 points btn -5 and 5
y_vals_1=slope*x_vals_1+intercept

plt.plot(x_vals_1,y_vals_1,color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='gray', linestyle='--') 
plt.axvline(0, color='gray', linestyle='--') 
plt.show()
```

```python
#cubic y=x^3-4x^2
x_vals_3=np.linspace(-5,5,500) #500 points between -5 and 5
y_vals_3=np.power((x_vals_3),3)-4*np.power(x_vals_3,2)
plt.plot(x_vals_3,y_vals_3,color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='gray', linestyle='--') 
plt.axvline(0, color='gray', linestyle='--') 
plt.show()
```
